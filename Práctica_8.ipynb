# **Práctica #8. Análisis de un Dataset con Seaborn y Visualización Avanzada**

```python
#Importación de las librerías.
import matplotlib.pyplot as plt
import seaborn as sns

#Carga del dataset.
df = sns.load_dataset('tips')

#Obtención del análisis preliminar.
print(df.head())
print(df.info())
print(df.describe())

#Verificación de valores nulos o datos faltantes.
print(df.isnull().sum())
```

# a) Mapa de Calor (Heatmap)

```python
#Filtrado de los datos.
variablesnumericas = df.select_dtypes(include=['number']).columns

#Creación de la matriz de correlación.
matrizcorrelacion = df[variablesnumericas].corr()

#Creación del gráfico.
sns.heatmap(matrizcorrelacion, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

#Personalización de etiquetas.
plt.title('Matriz de Correlación entre Variables')
plt.xlabel('Variable')
plt.ylabel('Variable')

#Impresión del gráfico.
plt.show()
```

# b) Diagrama de Violín

```python
#Creación del gráfico.
sns.violinplot(x='day', y='tip', data=df, palette='pastel')

#Personalización de etiquetas.
plt.title('Distribución de las Propinas por Día')
plt.xlabel('Día')
plt.ylabel('Propina')

#Impresión del gráfico.
plt.show()
```

# c) Gráfico de Dispersión (Scatter Plot)

```python
#Creación del gráfico.
sns.scatterplot(x='total_bill', y='tip', data=df, hue='day', palette='Set1')

#Personalización de etiquetas.
plt.title('Relación entre Total de la Cuenta y Propina')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')

#Impresión del gráfico.
plt.show()
```

# d) Gráfico de Barras (Bar Plot)

```python
#Creación del gráfico.
sns.barplot(x='sex', y='tip', data=df, estimator='mean', palette='bright')

#Personalización de etiquetas.
plt.title('Promedio de las Propinas por Género')
plt.xlabel('Género')
plt.ylabel('Propina Promedio')

#Impresión del gráfico.
plt.show()
```
