# Lectura de un archivo en Python
Para leer un archivo completo en Python, puedes usar varias formas, dependiendo de cómo quieras procesar el contenido. A continuación te muestro cómo hacerlo de las maneras más comunes:

**Este método lee todo el contenido del archivo y lo guarda como una cadena de texto:**

````python
# Abre el archivo en modo lectura ('r')
with open('/content/drive/MyDrive/ClassFiles/mbox-short.txt', 'r') as archivo:
    # Lee todo el contenido del archivo
    contenido = archivo.read()

# Imprime el contenido del archivo
print(contenido)
````

**Si prefieres leer el archivo línea por línea, puedes usar un bucle para iterar sobre las líneas del archivo:**

````python
with open('/content/drive/MyDrive/ClassFiles/mbox-short.txt', 'r') as archivo:
    # Itera sobre cada línea del archivo
    for linea in archivo:
        # Imprime cada línea
        print(linea, end='')  # end='' evita un salto de línea extra
````

**Si deseas almacenar todas las líneas en una lista, puedes hacerlo de la siguiente manera:**

````python
# Abre el archivo en modo lectura ('r')
with open('/content/drive/MyDrive/ClassFiles/mbox-short.txt', 'r') as archivo:
    # Lee todas las líneas y las guarda en una lista
    lineas = archivo.readlines()
# Imprime todas las líneas
print(lineas)
````

#Búsqueda a través de un archivo

Cuando buscas a través de los datos de un archivo, un patrón muy común es leer el archivo, ignorar la mayoría de las líneas y solamente procesar líneas que cumplan con una condición particular. Podemos combinar el patrón de leer un archivo con métodos de cadenas para construir mecanismos de búsqueda sencillos.


Por ejemplo, si queremos leer un archivo y solamente imprimir las líneas que comienzan con el prefijo “From:”, podríamos usar el método de cadenas startswith para seleccionar solo aquellas líneas con el prefijo deseado:

````python
archivo = open('/content/drive/MyDrive/ClassFiles/mbox.txt')
contador = 0
for linea in archivo:
  if linea.startswith('From:'):
    print(linea)
````

La salida parece correcta porque las líneas que estamos buscando son aquellas que comienzan con "From:", pero ¿por qué vemos líneas vacías adicionales? Esto ocurre debido al carácter invisible de salto de línea. Cada línea leída termina con un salto de línea, por lo que cuando usamos la sentencia print para imprimir el contenido de la variable `line`, esta incluye el salto de línea al final, y luego el comando print agrega otro salto de línea, lo que provoca el efecto de doble salto de línea que observamos.

Podemos usar el troceado de líneas para imprimir todos los caracteres excepto el último, pero una forma más sencilla de hacerlo es utilizar el método rstrip, que elimina los espacios en blanco al final de una cadena, como por ejemplo:

````python
archivo = open('/content/drive/MyDrive/ClassFiles/mbox.txt')
for linea in archivo:
    linea = linea.rstrip()
    if linea.startswith('From:'):
        print(linea)
````

A medida que tus programas para procesar archivos se vuelvan más complejos, puede que desees organizar tus bucles de búsqueda utilizando la instrucción continue. El concepto principal de un bucle de búsqueda es que estás buscando líneas "relevantes" e ignorando las "irrelevantes". Y cuando encuentras una línea relevante, realizas alguna acción con ella.

Podemos estructurar el bucle siguiendo el patrón de ignorar las líneas irrelevantes de la siguiente manera:

````python
archivo = open('/content/drive/MyDrive/ClassFiles/mbox.txt')
for linea in archivo:
    linea = linea.rstrip()
    # Ignorar 'líneas que no nos interesan'
    if not linea.startswith('From:'):
        continue
    # Procesar la línea que nos 'interesa'
    print(linea)
````

El resultado del programa es el mismo. En este caso, las líneas irrelevantes son aquellas que no comienzan con "From:", por lo que las ignoramos utilizando continue. Por otro lado, las líneas "relevantes" (las que comienzan con "From:") las procesamos.

Podemos usar el método find de las cadenas para simular la función de búsqueda de un editor de texto, que localiza las líneas donde aparece una cadena en cualquier parte. Dado que find busca cualquier ocurrencia de una cadena dentro de otra y devuelve la posición de esa cadena o -1 si no se encuentra, podemos escribir el siguiente bucle para mostrar las líneas que contienen la cadena "@uct.ac.za" (es decir, las provenientes de la Universidad de Cape Town en Sudáfrica):

````python
archivo = open('/content/drive/MyDrive/ClassFiles/mbox.txt')
for linea in archivo:
    linea = linea.rstrip()
    if linea.find('@uct.ac.za') == -1: continue
    print(linea)
````

# Cerrar un archivo de forma correcta

En Python, es importante cerrar los archivos después de usarlos para evitar pérdidas de datos o errores, especialmente si el archivo ha sido abierto en modo de escritura o se han realizado cambios. Hay dos formas principales de hacerlo:
************
El uso de with es la manera más sencilla y segura de manejar archivos en Python, ya que se encarga automáticamente de cerrar el archivo una vez que el bloque de código se completa, incluso si ocurre un error dentro del bloque.

```
# Abrir el archivo con 'with', lo que asegura que se cierre automáticamente
with open('archivo.txt', 'r') as archivo:
    # Leer el contenido o procesar el archivo
    contenido = archivo.read()
    print(contenido)

# No es necesario llamar a archivo.close(), ya que 'with' lo hace automáticamente.
```
***********
Usando open() y close() manualmente
Si no usas with, debes asegurarte de cerrar el archivo de forma explícita después de haber terminado de trabajar con él. Esto se hace con el método close().



```
# Abrir el archivo manualmente
archivo = open('archivo.txt', 'r')

# Leer el contenido o procesar el archivo
contenido = archivo.read()
print(contenido)

# Cerrar el archivo manualmente
archivo.close()
```
