Ejercicio 1: Normalización Min-Max

1.	Cargar el dataset wine de sklearn.
2.	Seleccionar una de las características numéricas (por ejemplo, alcohol).
3.	Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.
4.	Aplicar la normalización Min-Max a los datos seleccionados.
5.	Realizar la prueba de normalidad nuevamente sobre los datos normalizados.
6.	Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.
7.	Responder las preguntas siguientes.

Preguntas:
1. ¿Cómo afectó la normalización Min-Max a la distribución de los datos? La distribución de los datos no se vio afectada, sin embargo, la escala de los valores de los datos sí que tuvo un cambio, originalmente [11, 15], ahora [0, 1].
2. ¿Qué interpretas del p-valor antes y después de la normalización? El p-valor se mantuvo igual para ambos casos, esto quiere decir que los datos mantuvieron su distribución original después de la normalización.


```python
#Importación de librerías.
from sklearn.datasets import load_wine
import pandas as pd
import scipy.stats as stats
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

#Carga de dataset.
df = pd.DataFrame(load_wine().data, columns=load_wine().feature_names)

#Selección de característica.
seleccion = df[['alcohol']]

#Cálculo de prueba de normalidad.
prueba_shapiro = stats.shapiro(seleccion)
print(f"Prueba Shapiro-Wilk sobre datos originales: p-valor = {prueba_shapiro.pvalue:.5f}")

#Aplicación de normalización.
escalador = MinMaxScaler(feature_range=(0, 1))
seleccion_escalada = escalador.fit_transform(seleccion)

#Cálculo de prueba de normalidad.
prueba_shapiro_escalada = stats.shapiro(seleccion_escalada)
print(f"Prueba Shapiro-Wilk sobre datos normalizados: p-valor = {prueba_shapiro_escalada.pvalue:.5f}")

#Creación y diseño de gráficos.
plt.figure(figsize=(8, 4))

#Gráfico de datos originales.
plt.subplot(1, 2, 1)
plt.scatter(range(len(seleccion)), seleccion, color='red')
plt.title("Datos Originales")

#Gráfico de datos normalizados.
plt.subplot(1, 2, 2)
plt.scatter(range(len(seleccion_escalada)), seleccion_escalada, color='blue')
plt.title("Datos Normalizados [0, 1]")

#Impresión de gráficos.
plt.show()
```

# Ejercicio 2: Normalización Z-Score

Instrucciones:
1.	Cargar el dataset wine de sklearn.
2.	Seleccionar una de las características numéricas (por ejemplo, malic_acid).
3.	Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.
4.	Aplicar la normalización Z-Score a los datos seleccionados.
5.	Realizar la prueba de normalidad nuevamente sobre los datos normalizados.
6.	Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.
7.	Responder las preguntas siguientes.

Preguntas:
1.	¿Qué diferencia observas entre la distribución original y la normalizada con Z-Score? La distribución de los datos no se vio afectada, sin embargo, la escala de los valores de los datos sí que tuvo un cambio, originalmente [0, 6], ahora [-2, 4] aproximadamente.
2.	¿Qué implicaciones tiene la normalización Z-Score en el análisis de los datos? Permite que los datos tengan una escala comparable, sin importar las unidades originales.



```python
#Importación de librerías.
from sklearn.datasets import load_wine
import pandas as pd
import scipy.stats as stats
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

#Carga de dataset.
df = pd.DataFrame(load_wine().data, columns=load_wine().feature_names)

#Selección de característica.
seleccion = df[['malic_acid']]

#Cálculo de prueba de normalidad.
prueba_shapiro = stats.shapiro(seleccion)
print(f"Prueba Shapiro-Wilk sobre datos originales: p-valor = {prueba_shapiro.pvalue:.5f}")

#Aplicación de normalización.
escalador = StandardScaler()
seleccion_estandarizada = escalador.fit_transform(seleccion)

#Cálculo de prueba de normalidad.
prueba_shapiro_estandarizada = stats.shapiro(seleccion_estandarizada)
print(f"Prueba Shapiro-Wilk sobre datos estandarizados: p-valor = {prueba_shapiro_estandarizada.pvalue:.5f}")

#Creación y diseño de gráficos.
plt.figure(figsize=(8, 4))

#Gráfico de datos originales.
plt.subplot(1, 2, 1)
plt.scatter(range(len(seleccion)), seleccion, color='red')
plt.title("Datos Originales")

#Gráfico de datos estandarizados.
plt.subplot(1, 2, 2)
plt.scatter(range(len(seleccion_estandarizada)), seleccion_estandarizada, color='blue')
plt.title("Datos Estandarizados")

#Impresión de gráficos.
plt.show()
```

Ejercicio 3: Transformación Logarítmica
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar una transformación logarítmica (np.log) a los datos.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación logarítmica a la distribución de los datos? Luego de la transformación logarítmica, los datos se ajustaron y parece ser que se observan más lineales de lo que eran antes.
2.	¿Qué interpretas del p-valor antes y después de la transformación logarítmica? Desafortunadamente, el p-valor para ambos casos es el mismo, con un valor de 0, lo que sugiere que la distribución de los datos no es normal. Esto puede deberse a varios factores, como lo es el hecho de generar de manera aleatoria el conjunto de datos.


```python
#Importación de librerías.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#Carga de conjunto de datos.
x = np.random.randint(1, 101, 100)
y = x ** 2

#Aplicación de transformación.
y_transformada = np.log(y)

#Cálculo de pruebas de normalidad.
prueba_shapiro = stats.shapiro(y)
print(f"Prueba Shapiro-Wilk sobre datos originales: p-valor = {prueba_shapiro.pvalue:.5f}")
prueba_shapiro_transformada = stats.shapiro(y_transformada)
print(f"Prueba Shapiro-Wilk sobre datos transformados: p-valor = {prueba_shapiro_transformada.pvalue:.5f}")

#Creación y diseño de gráficos.
plt.figure(figsize=(12, 6))

#Gráfico de datos originales.
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='red')
plt.title("Datos Originales")
plt.xlabel("x")
plt.ylabel("y")

#Gráfico de datos transformados.
plt.subplot(1, 2, 2)
plt.scatter(x, y_transformada, color='blue')
plt.title("Datos Transformados")
plt.xlabel("x")
plt.ylabel("log (y)")

#Impresión de gráficos.
plt.show()
```

Ejercicio 4: Transformación Raíz Cuadrada
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar una transformación de raíz cuadrada a los datos.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación raíz cuadrada a la distribución de los datos? Ha comprimido el rango de los datos, originalmente [0, 10000], ahora [0, 100]. Sin embargo, debido a la naturaleza de los datos, es posible que no presenten una distribución normal, pero si una relación más lineal.
2.	¿Qué efectos tiene la raíz cuadrada sobre los valores extremos de los datos? Los comprime, y reduce la dimensionalidad del rango que abarca el conjunto de datos.


```python
#Importación de librerías.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#Carga de conjunto de datos.
x = np.random.randint(1, 101, 100)
y = x ** 2

#Aplicación de transformación.
y_transformada = np.sqrt(y)

#Cálculo de pruebas de normalidad.
prueba_shapiro = stats.shapiro(y)
print(f"Prueba Shapiro-Wilk sobre datos originales: p-valor = {prueba_shapiro.pvalue:.5f}")
prueba_shapiro_transformada = stats.shapiro(y_transformada)
print(f"Prueba Shapiro-Wilk sobre datos transformados: p-valor = {prueba_shapiro_transformada.pvalue:.5f}")

#Creación y diseño de gráficos.
plt.figure(figsize=(12, 6))

#Gráfico de datos originales.
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='red')
plt.title("Datos Originales")
plt.xlabel("x")
plt.ylabel("y")

#Gráfico de datos transformados.
plt.subplot(1, 2, 2)
plt.scatter(x, y_transformada, color='blue')
plt.title("Datos Transformados")
plt.xlabel("x")
plt.ylabel("sqrt (y)")

#Impresión de gráficos.
plt.show()
```

Ejercicio 5: Transformación Box-Cox
Instrucciones:
1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar la transformación Box-Cox a los datos utilizando scipy.stats.boxcox.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación Box-Cox a la distribución de los datos? La transformación hizo que los datos fueran más lineales respecto a los datos originales.
2.	¿Qué interpretación tiene el valor lambda obtenido en la transformación Box-Cox? Es el valor que determinara el tipo de transformación que se aplicara, para este caso, λ = 0.41, por lo que se acerca a una transformación del tipo raíz cuadrada.


```python
#Importación de librerías.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

#Carga de conjunto de datos.
x = np.random.randint(1, 101, 100)
y = x ** 2

#Aplicación de transformación.
y_transformada, lambda_optimo = stats.boxcox(y)

#Cálculo de pruebas de normalidad.
prueba_shapiro = stats.shapiro(y)
print(f"Prueba Shapiro-Wilk sobre datos originales: p-valor = {prueba_shapiro.pvalue:.5f}")
prueba_shapiro_transformada = stats.shapiro(y_transformada)
print(f"Prueba Shapiro-Wilk sobre datos transformados: p-valor = {prueba_shapiro_transformada.pvalue:.5f}")

#Creación y diseño de gráficos.
plt.figure(figsize=(12, 6))

#Gráfico de datos originales.
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='red')
plt.title("Datos Originales")
plt.xlabel("x")
plt.ylabel("y")

#Gráfico de datos transformados.
plt.subplot(1, 2, 2)
plt.scatter(x, y_transformada, color='blue')
plt.title(f"Datos Transformados (λ óptimo = {lambda_optimo:.2f})")
plt.xlabel("x")
plt.ylabel("Box-Cox (y)")

#Impresión de gráficos.
plt.show()
```
