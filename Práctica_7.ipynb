# **Práctica #7. Análisis de un Dataset con Matplotlib**

```python
#Importación de las librerías.
import matplotlib.pyplot as plt
import seaborn as sns

#Carga del dataset.
df = sns.load_dataset('iris')

#Verificación de las primeras líneas.
print(df.head())

#Obtención de información general sobre los datos.
print(df.info())
```

# a) Gráfico de barras

```python
#Importación de la librería NumPy.
import numpy as np

#Cálculo de las medias.
largo = df.groupby('species')['sepal_length'].mean()
ancho = df.groupby('species')['sepal_width'].mean()

#Creación del gráfico.
barraposicion = np.arange(len(largo))
barraanchura = 0.35
plt.bar(barraposicion - barraanchura/2, largo.values, barraanchura, color='red', edgecolor='black', label='Largo')
plt.bar(barraposicion + barraanchura/2, ancho.values, barraanchura, color='yellow', edgecolor='black', label='Ancho')
plt.xticks(barraposicion, largo.index)
plt.legend()

#Personalización de etiquetas.
plt.title('Promedios del Largo y Ancho por Especie')
plt.xlabel('Especie')
plt.ylabel('Media')

#Impresión del gráfico.
plt.show()
```

# b) Histograma

```python
#Creación del gráfico.
plt.hist(df['sepal_length'], bins=20, color='blue', edgecolor='black')

#Personalización de etiquetas.
plt.title('Distribución del Largo de los Sépalos')
plt.xlabel('Largo')
plt.ylabel('Frecuencia')

#Impresión del gráfico.
plt.show()
```

# c) Gráfico de dispersión (scatter plot)

```python
#Creación del gráfico.
plt.scatter(df[df['species'] == 'setosa']['petal_length'], df[df['species'] == 'setosa']['petal_width'], alpha=0.5, color='orange', label='setosa')
plt.scatter(df[df['species'] == 'versicolor']['petal_length'], df[df['species'] == 'versicolor']['petal_width'], alpha=0.5, color='green', label='versicolor')
plt.scatter(df[df['species'] == 'virginica']['petal_length'], df[df['species'] == 'virginica']['petal_width'], alpha=0.5, color='purple', label='virginica')
plt.legend()

#Personalización de etiquetas.
plt.title('Relación entre Largo y Ancho de los Pétalos')
plt.xlabel('Largo')
plt.ylabel('Ancho')

#Impresión del gráfico.
plt.show()
```

# d) Gráfico de cajas (box plot)

```python
#Filtrado de los datos para el largo.
largosetosa = df[df['species'] == 'setosa']['sepal_length']
largoversicolor = df[df['species'] == 'versicolor']['sepal_length']
largovirginica = df[df['species'] == 'virginica']['sepal_length']

#Filtrado de los datos para el ancho.
anchosetosa = df[df['species'] == 'setosa']['sepal_width']
anchoversicolor = df[df['species'] == 'versicolor']['sepal_width']
anchovirginica = df[df['species'] == 'virginica']['sepal_width']

#Creación del gráfico.
plt.figure(figsize=(12, 6))

#Creación y personalización del subgráfico para el largo.
plt.subplot(1, 2, 1)
plt.boxplot([largosetosa, largoversicolor, largovirginica], labels=['setosa', 'versicolor', 'virginica'])
plt.title('Distribución del Largo del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Largo')

#Creación y personalización del subgráfico para el ancho.
plt.subplot(1, 2, 2)
plt.boxplot([anchosetosa, anchoversicolor, anchovirginica], labels=['setosa', 'versicolor', 'virginica'])
plt.title('Distribución del Ancho del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Ancho')

#Ajuste del tamaño del gráfico.
plt.tight_layout()

#Impresión del gráfico.
plt.show()
```

# e) Gráfico de líneas

```python
#Creación del gráfico.
plt.plot(df.index, df['petal_length'], color='red', label='Longitud', linewidth=1)

#Personalización de etiquetas.
plt.title('Tendencia de la Longitud del Pétalo')
plt.xlabel('Índice de Observación')
plt.ylabel('Longitud')

#Impresión del gráfico.
plt.show()
```
